```python
import pandas as pd
```

```python
df = pd.read_csv('financial_data.csv')
df
```

#### Sort Data

```python
df = df.sort_values(by=['Company', 'Fiscal Year'])
df
```

#### Growth Calculation 

```python
df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue ($M)'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income ($M)'].pct_change() * 100
df['Asset Growth (%)'] = df.groupby('Company')['Total Assets ($M)'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby('Company')['Operating Cash Flow'].pct_change() * 100

df
```

```python
growth_cols = ['Revenue Growth (%)', 'Net Income Growth (%)']
df[growth_cols] = df[growth_cols].fillna(0)
print("📊 Financial Data with Growth Metrics:\n")
print(df)
```

#### Profit Margin

```python
df['Profit Margin (%)'] = (df['Net Income ($M)'] / df['Total Revenue ($M)']) * 100
df
```

#### Company-wise Summary

```python
df.groupby('Company').mean(numeric_only=True)
```

```python
summary = df.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean'
}).reset_index()


print("\nYear-over-Year Average Growth Rates (%):")
print(summary)
```

### 📊 Financial Data Analysis Summary
🔍 Overview

This analysis evaluates the financial performance of Microsoft, Apple, and Tesla over the last three fiscal years using key financial metrics such as revenue, net income, assets, liabilities, and operating cash flow. Growth rates and profitability ratios were calculated to identify trends and insights relevant for AI-driven financial analysis.

🏢 Company-wise Insights
Microsoft
Microsoft demonstrates strong and consistent revenue growth (~15% YoY in recent years).

Net income has shown steady growth, indicating strong profitability.
Significant growth in assets reflects business expansion and investment capacity.
Operating cash flow has increased consistently, highlighting excellent liquidity and operational efficiency.
Profit margins remain high (~34–36%), indicating a highly profitable and scalable business model.

👉 Conclusion:
Microsoft is financially robust with strong growth, high profitability, and efficient cash generation, making it a stable and high-performing company.

Apple
Apple shows moderate but stable revenue growth (~2%–6%)

Net income declined slightly in 2024 but recovered strongly in 2025, indicating resilience.
Asset growth is relatively stable, with slight fluctuations.
Operating cash flow remains strong but shows minor volatility.
Profit margins are consistently healthy (~24–27%).

👉 Conclusion:
Apple maintains a strong financial position with stable growth and profitability, though slight fluctuations suggest sensitivity to costs and external factors.

Tesla
Tesla’s revenue growth is stagnant to slightly declining in recent years.
Net income shows a sharp decline (-52% and -46%), indicating significant profitability challenges.
Assets are increasing, suggesting continued investment and expansion.
Operating cash flow is relatively stable but not growing significantly.
Profit margins have dropped drastically (~15% → ~4%), indicating margin pressure.

👉 Conclusion:
Tesla is facing declining profitability despite asset growth, suggesting rising costs, pricing pressure, or operational inefficiencies.

📈 Key Comparative Insights
Microsoft leads in profitability and consistent growth.
Apple shows stability and resilience with strong margins.
Tesla shows declining profitability, making it comparatively riskier.
