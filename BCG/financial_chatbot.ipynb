```python
import pandas as pd

# Load your dataset
df = pd.read_csv('financial_data.csv')
df = df.sort_values(by=['Company', 'Fiscal Year'])
```

#### Chatbot Function

```python
def chatbot(user_query):
    user_query = user_query.lower()

    # Microsoft Queries
    if "microsoft" in user_query:
        if "revenue" in user_query:
            revenue = df[(df['Company'] == 'Microsoft') & (df['Fiscal Year'] == 2025)]['Total Revenue ($M)'].values[0]
            return f"Microsoft's total revenue in 2025 was {revenue} million USD, showing strong growth."
        
        elif "growth" in user_query:
            growth = df[(df['Company'] == 'Microsoft') & (df['Fiscal Year'] == 2025)]['Revenue Growth (%)'].values[0]
            return f"Microsoft's revenue grew by {growth:.2f}% in 2025, indicating consistent expansion."
        
        elif "profit" in user_query:
            margin = df[(df['Company'] == 'Microsoft') & (df['Fiscal Year'] == 2025)]['Profit Margin (%)'].values[0]
            return f"Microsoft has a strong profit margin of {margin:.2f}%, reflecting high profitability."

    # Apple Queries
    elif "apple" in user_query:
        if "revenue" in user_query:
            revenue = df[(df['Company'] == 'Apple') & (df['Fiscal Year'] == 2025)]['Total Revenue ($M)'].values[0]
            return f"Apple's total revenue in 2025 was {revenue} million USD, showing steady growth."
        
        elif "income" in user_query:
            income = df[(df['Company'] == 'Apple') & (df['Fiscal Year'] == 2025)]['Net Income ($M)'].values[0]
            return f"Apple's net income in 2025 was {income} million USD, indicating strong profitability."
        
        elif "profit" in user_query:
            margin = df[(df['Company'] == 'Apple') & (df['Fiscal Year'] == 2025)]['Profit Margin (%)'].values[0]
            return f"Apple maintains a solid profit margin of {margin:.2f}%."

    # Tesla Queries
    elif "tesla" in user_query:
        if "revenue" in user_query:
            revenue = df[(df['Company'] == 'Tesla') & (df['Fiscal Year'] == 2025)]['Total Revenue ($M)'].values[0]
            return f"Tesla's total revenue in 2025 was {revenue} million USD."
        
        elif "growth" in user_query:
            growth = df[(df['Company'] == 'Tesla') & (df['Fiscal Year'] == 2025)]['Revenue Growth (%)'].values[0]
            return f"Tesla's revenue growth in 2025 was {growth:.2f}%, indicating slow or negative growth."
        
        else:
            margin = df[(df['Company'] == 'Tesla') & (df['Fiscal Year'] == 2025)]['Profit Margin (%)'].values[0]
            return f"Tesla's profit margin is {margin:.2f}%, showing declining profitability."

    # Best Company Query
    elif "best" in user_query or "highest profit" in user_query:
        best = df.loc[df['Profit Margin (%)'].idxmax()]
        return f"{best['Company']} has the highest profit margin at {best['Profit Margin (%)']:.2f}%."

    # Default Response
    else:
        return "Sorry, I can answer questions about Microsoft, Apple, and Tesla financial performance."
```

#### Run Chatbot Loop

```python
print("💬 Financial Chatbot (type 'exit' to quit)")

while True:
    query = input("You: ")
    
    if query.lower() == "exit":
        print("Chatbot: Goodbye!")
        break
    
    print("Chatbot:", chatbot(query))
```
